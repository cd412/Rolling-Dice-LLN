# Rolling Dice and the Law of Large Numbers

#### Interview Question
Here's a challenge: Please write a program that simulates and outputs the result of 100 rolls of a fair 20-sided die. Paste your code below. What is your estimate for the average of those rolls (including standard deviation)? What is your estimate for the expected value of a single roll of a 20-sided die (including standard deviation)? Briefly explain why the two estimates are or are not the same.*

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
def roll_dice(n_trials, n_rolls_per_trial, n_sides_of_die=20):
    results = np.zeros(n_trials)
    for i in range(n_trials):
        results[i] = np.random.randint(1, n_sides_of_die+1, n_rolls_per_trial).mean()
    print('MEAN: {}\nSTD: {}'.format(results.mean(), results.std())) 
    plt.hist(results)
```

```python
# Estimated Mean and Standard Deviation for 1 roll
roll_dice(100000, 1)

# Mean = (a+b)/2 = (1+20)/2 = 10.5
# Std = sqrt(((b-a+1)^2-1)/12) = sqrt(399/12) = 5.76628
```

```python
# Estimated Mean and Standard Deviation for average value of 100 rolls
roll_dice(100000, 100)
```

## Conclusion
The estimates differ because the experiments changed.

In the single-roll experiment, this is a discrete uniform distribution. Each value has an equal chance of being rolled. The standard deviation is high.

In the 100-roll experiment, the average of the rolls are not equally distributed. For example, to get an averge roll of 1 requires rolling 1 100 times with the slim probability of (1/20)^100. 

This is an example of the Law of Large Numbers.

Intuitively, it becomes more likely for the average of the rolls to be near the expected value as the number of rolls per trial increases. That is why the standard deviation shrinks as n_rolls_per_trial increases.

```python
roll_dice(100000, 1)
roll_dice(100000, 10)
roll_dice(100000, 100)
roll_dice(100000, 1000)
plt.gca().set_title('Standard Deviation Decreases as n_rolls_per_trial Increases')
```
